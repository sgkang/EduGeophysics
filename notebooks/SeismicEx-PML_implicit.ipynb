## Acoustic Wave equation in time domain (1st order form)

$ \rho\frac{\partial p}{\partial t} = \nabla \cdot \vec{u}+\frac{\partial s}{\partial t}\delta(\vec{r}-\vec{r}_s)$

$ \mu^{-1}\frac{\partial \vec{u}}{\partial t} = \nabla p$

## Add damping term

$ \rho[\frac{\partial p}{\partial t} + \sigma p] = \nabla \cdot \vec{u}+\frac{\partial s}{\partial t}\delta(\vec{r}-\vec{r}_s)$

$ \mu^{-1}[\frac{\partial \vec{u}}{\partial t}+\sigma \vec{u}] = \nabla p$

## Discretized form

$ \mathbf{diag}(\rho)\frac{\mathbf{p}^{n+1}-\mathbf{p}^{n}}{\triangle t} +  \mathbf{diag}(\rho\sigma)\mathbf{p}^n = \mathbf{Div} \mathbf{u}^{n}+\mathbf{M}^{cc -1}\frac{\mathbf{s}^{n+1}-\mathbf{s}^{n}}{\triangle t}$ 


$ \mathbf{diag}(\mathbf{Av}^{f}_{cc}\mu^{-1})\frac{\mathbf{u}^{n+1}-\mathbf{u}^{n}}{\triangle t} + \mathbf{diag}(\mathbf{Av}^{f}_{cc}\mu^{-1}\sigma)\mathbf{u}^{n+1}=  \mathbf{Grad} \mathbf{p}^{n+1}$

<center>

\begin{eqnarray}
  \begin{bmatrix}
    \triangle t^{-1}\mathbf{diag}(\rho) & \tilde{\mathbf{0}} \\[0.3em]
    -\mathbf{Grad} & \triangle t^{-1}\mathbf{diag}(\mathbf{Av}^{f}_{cc}\mu^{-1}) + \mathbf{diag}(\mathbf{Av}^{f}_{cc}\mu^{-1}\sigma)          \\[0.3em]
  \end{bmatrix}
  \begin{bmatrix}
     \mathbf{p}^{n+1} \\[0.3em]
     \mathbf{u}^{n+1} \\[0.3em]
  \end{bmatrix} + 
  \begin{bmatrix}
    -\triangle t^{-1}\mathbf{diag}(\rho)+ \mathbf{diag}(\rho\sigma)& -\mathbf{Div} \\[0.3em]
    \tilde{\mathbf{0}} & -\triangle t^{-1}\mathbf{diag}(\mathbf{Av}^{f}_{cc}\mu^{-1})          \\[0.3em]
  \end{bmatrix}
  \begin{bmatrix}
     \mathbf{p}^{n} \\[0.3em]
     \mathbf{u}^{n} \\[0.3em]
  \end{bmatrix} =   
  \begin{bmatrix}
     \triangle t^{-1}\mathbf{M}^{cc -1}(\mathbf{s}^{n+1}-\mathbf{s}^{n}) \\[0.3em]
     \tilde{\mathbf{0}} \\[0.3em]
  \end{bmatrix}
\end{eqnarray}

```python
mesh.setCellGradBC('dirichlet')
Grad = mesh.cellGrad
Div = mesh.faceDiv
rho = np.ones(mesh.nC)*2.7
vhalf = 2000
vblk = 2800
v = np.ones(mesh.nC)*vhalf
blkind = np.logical_and(mesh.gridCC[:,1]>-12.5, mesh.gridCC[:,1]<12.5) & np.logical_and(mesh.gridCC[:,0]>-12.5, mesh.gridCC[:,0]<12.5)
v[blkind] = vblk
mu = rho*v**2
AvF2CC = mesh.aveF2CC
```

```python
time = np.linspace(0, 0.08, 2**10)
dt = time[1]-time[0]
```

```python
ax = mesh.vectorCCx[-30]
ay = mesh.vectorCCy[-30]
indy = np.logical_or(mesh.gridCC[:,1]<=-ay, mesh.gridCC[:,1]>=ay)
indx = np.logical_or(mesh.gridCC[:,0]<=-ax, mesh.gridCC[:,0]>=ax)

sigmax = 100
sigx = zeros_like(mesh.gridCC[:,0])
sigx[indx] = (abs(mesh.gridCC[:,0][indx])-ax)**2
sigx[indx] = sigx[indx]-sigx[indx].min()
sigx[indx] = sigx[indx]/sigx[indx].max()
```

```python
sigy = zeros_like(mesh.gridCC[:,1])
sigy[indy] = (abs(mesh.gridCC[:,1][indy])-ay)**2
sigy[indy] = sigy[indy]-sigy[indy].min()
sigy[indy] = sigy[indy]/sigy[indy].max()
```

```python
L = (mesh.hx.sum()-ax)*0.5
sigmax = L*100
```

```python
sig = (sigx+sigy)*sigmax
sig[indy & indx] = sig[indy & indx]*0.5
```

```python
plt.plot(mesh.vectorCCx, sig[mesh.gridCC[:,1]==mesh.vectorCCy[125]], '.')
```

```python
sigy = zeros_like(mesh.gridCC[:,1])
sigy[mesh.gridCC[:,1]<-ay] = (mesh.gridCC[:,1][mesh.gridCC[:,1]<-ay]-ay)**2
refy = sigy[mesh.gridCC[:,0]<-ay].max()
sigy[mesh.gridCC[:,1]>ay] = (mesh.gridCC[:,1][mesh.gridCC[:,1]>ay]+ay)**2
sigy = sigy/refy
```

```python
An = sp.vstack((sp.hstack((1/dt*Utils.sdiag(rho), Utils.spzeros(mesh.nC, mesh.nF))),\
                sp.hstack((-Grad, 1/dt*Utils.sdiag(AvF2CC.T*(1/mu)) + Utils.sdiag(AvF2CC.T*(1/mu*sig))))))
Bn = sp.vstack((sp.hstack((-1/dt*Utils.sdiag(rho)+Utils.sdiag(rho*sig), -Div)),\
                sp.hstack((Utils.spzeros(mesh.nF, mesh.nC), -1/dt*Utils.sdiag(AvF2CC.T*(1/mu))))))
```

```python
fig, ax = plt.subplots(1,2, figsize = (10, 5))
ax[0].spy(An, ms = 1)
ax[0].grid(True)
ax[1].spy(Bn, ms = 1)
ax[1].grid(True)
```

```python
def ricker(fpeak, t, tlag):
    """
        Generating Ricker Wavelet
        
        .. math ::
        
        
    """
    return (1-2*np.pi**2*fpeak**2*(t-tlag)**2)*np.exp(-np.pi**2*fpeak**2*(t-tlag)**2)
```

```python
wave = ricker(600, time, 0.0025)
```

```python
Proj = mesh.getInterpolationMat(np.r_[3., 30], 'CC')
```

```python
%%time
b0 = np.zeros(mesh.nC+mesh.nF)
for i in range(time.size-1):
    s1 = ricker(400, time[i+1], 0.0025)
    s0 = ricker(400, time[i], 0.0025)
    s = np.r_[q*(s1-s0)*1/dt, np.zeros(mesh.nF)]
    bn = Ainv*(s-Bn*b0)    
    p[:,i+1] = bn[0:mesh.nC]
    b0 = bn.copy()
```
