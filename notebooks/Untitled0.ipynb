```python
time = np.linspace(0, 0.8, 2**10)
options={'tlag':0.0025, 'fmain':400}
rx = AcousticRx(np.vstack((np.r_[0, 1], np.r_[0, 1])))
tx1 = AcousticTx(np.r_[-15., 0], time, [rx], **options)
tx2 = AcousticTx(np.r_[0., 0.], time, [rx], **options)
tx3 = AcousticTx(np.r_[15., 0.], time, [rx], **options)
# survey = SurveyAcoustic([tx1, tx2, tx3])
survey = SurveyAcoustic([tx2])
cs = 0.3
hx = np.ones(200)*cs
hy = np.ones(200)*cs
mesh = Mesh.TensorMesh([hx, hy], 'CC')
prob = AcousticProblem(mesh)
sig = np.zeros(mesh.nC)
vhalf = 2000
v = np.ones(mesh.nC)*vhalf
prob.setAbsorbingBC(30)
prob.storefield = True
prob.pair(survey)
```

```python
prob.stabilitycheck(v, time, options['fmain'])
print prob.G
```
